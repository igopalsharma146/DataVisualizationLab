```python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
data = sns.load_dataset('iris')

# Filtering data
filtered_data = data[data['sepal_length'] > 5.0]

# Scatter plot for filtered data
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=filtered_data)
plt.title('Filtered Data: Sepal Length > 5.0')
plt.show()
```
