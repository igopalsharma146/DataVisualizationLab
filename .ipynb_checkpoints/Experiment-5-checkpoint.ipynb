```python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
data = sns.load_dataset('iris')

# Apply log transformation
data['log_sepal_length'] = np.log(data['sepal_length'])

# Plot transformed data
sns.histplot(data['log_sepal_length'], bins=10)
plt.title('Log Transformed Sepal Length Histogram')
plt.show()
```
