```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
data = sns.load_dataset('iris')

# Create another data source
extra_data = pd.DataFrame({'extra_feature': [1, 2, 3, 4, 5]})

# Combine datasets
combined_data = pd.concat([data.head(), extra_data], axis=1)

# Scatter plot using combined data
sns.scatterplot(x='sepal_length', y='extra_feature', data=combined_data)
plt.title('Combined Data Visualization')
plt.show()
```
