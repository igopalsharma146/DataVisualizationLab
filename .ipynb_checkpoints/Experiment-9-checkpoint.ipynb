```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt

# Generate skewed data
skewed_data = np.random.chisquare(df=4, size=1000)

# Plot original skewed data
plt.hist(skewed_data, bins=20)
plt.title('Skewed Data (Chi-Square Distribution)')
plt.show()

```

```python
# Apply log transformation to reduce skewness
log_transformed_data = np.log(skewed_data)

# Plot transformed data
plt.hist(log_transformed_data, bins=20)
plt.title('Log Transformed Data (Reduced Skewness)')
plt.show()
```
