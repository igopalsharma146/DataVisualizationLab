```python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
data = sns.load_dataset('iris')

# Scatter plot
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=data)
plt.title('Seaborn Scatter Plot')
plt.show()

```

```python
# Line chart
plt.plot([1, 2, 3, 4], [10, 15, 13, 17])
plt.title('Matplotlib Line Chart')
plt.show()

```

```python
# Bar chart
data['species'].value_counts().plot(kind='bar')
plt.title('Matplotlib Bar Chart')
plt.show()

```

```python
# Histogram
data['sepal_length'].plot(kind='hist', bins=10)
plt.title('Matplotlib Histogram')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
data = sns.load_dataset('iris')

# Exclude the 'species' column and compute the correlation matrix
corr = data.drop('species', axis=1).corr()

# Create a heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')

# Display the plot
plt.show()
```
