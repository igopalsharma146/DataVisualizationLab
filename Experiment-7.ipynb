```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Simulate sales data
dates = pd.date_range('2023-01-01', periods=100)
sales_data = pd.DataFrame({'Date': dates, 'Sales': np.random.randint(100, 500, size=(100))})

# Time series plot
plt.plot(sales_data['Date'], sales_data['Sales'])
plt.title('Sales Time Series')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```
